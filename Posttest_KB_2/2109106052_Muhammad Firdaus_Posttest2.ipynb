pd.read_csv untuk mengakses file dari directory yang akan digunakan.

```python
dataset=pd.read_csv("tsunami_dataset.csv")
```

memasukkan .shape kedalam 2 variabel yaitu record untuk baris dan attribut untuk kolom. memasukan nama kolom ke dalam variabel. memasukkan jenis data angka ke jumlah_atribut_angka dan memasukkan nilai null ke jumlah_atribut_null.

```python
record,attribute=dataset.shape
kolom=dataset.columns.values
jumlah_atribut_angka=dataset.select_dtypes(["int","float"]).any().sum()
jumlah_atribut_null=dataset.isnull().any().sum()
```

menampilkan semua info secara satu-persatu dengan cara memasukkannya di dalam dictionary{}

```python
info={"record":record,"kolom":attribute,"nama kolom":list(kolom),"jumlah angka":jumlah_atribut_angka,"jumlah null":jumlah_atribut_null}
```

```python
for x in info:
    print(x,":",info[x])
```
