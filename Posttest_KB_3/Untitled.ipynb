```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

import library yang dibutuhkan untuk melakukan visualisasi data.

```python
df = pd.read_csv("tsunami_dataset.csv")

df.head(20)
```

memperlihatkan table dari dataset yang telah disiapkan dimulai dari data ke 20 keatas.

```python
df['YEAR'] = pd.cut(df['YEAR'], bins = [-2000,-1500,-1000,-500,0,500,1000,1500,2000], labels=['sangat_lama_sekali','sangat_lama','lama_sekali','lama','baru','baru_sekali','sangat_baru','sangat_baru_sekali'])

df.head()
```

mengubah nilai dari YEAR menjadi lebih gampang untuk di data.

```python
df = pd.read_csv("tsunami_dataset.csv", on_bad_lines="skip")
df.info()
```

on_bad_line="skip" : mengabaikan line yang rusak atau error dalam data set.

```python
plt.figure(figsize=(25,10))
sns.countplot(data = df, x = 'CAUSE', hue='DAMAGE_TOTAL_DESCRIPTION')
plt.title("Perbandingan Pendapatan total kerusakan pada bencana alam")
plt.ylabel("Banyaknya kerusakan")
plt.xlabel("Bencana alam")
plt.show()
```

dari visualisasi data diatas dapat diketahui perbandingan dari bencana alam dan total kerusakan yang dialami. dan bencana alam yang menyebabkan total kerusakan yang banyak adalah gempa bumi.


```python
data_pie = df['DEATHS_TOTAL_DESCRIPTION'].value_counts().rename('DEATHS_TOTAL_DESCRIPTION').reset_index(name = 'id_count')
plt.pie(data_pie.id_count, labels = data_pie['index'], startangle = 180, autopct = '%.1f%%',explode = (0.1, 0, 0, 0),
       shadow = True)
plt.title("Banyaknya korban jiwa akibat tsunami", size = 10)
plt.show()
```

### piechart
pendataan berapa banyak korban jiwa akibat dari tsunami yang terjadi selama ini.

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,10))
sns.scatterplot(data=df,x='EVENT_VALIDITY',y='CAUSE',hue='DEATHS_TOTAL_DESCRIPTION')
plt.show()
```

pendataan visualisasi dari seberapa parah tsunami yang terjadi dikarena beberapa bencana alam dan banyaknya korban jiwa.

```python
plt.figure(figsize=(12,10))
sns.countplot(data = df, x = 'YEAR', hue='DAMAGE_TOTAL_DESCRIPTION')
plt.title("Perbandingan kerusakan di setiap tahun karena bencana alam")
plt.ylabel("Banyaknya kerusakan")
plt.xlabel("Bencana alam")
plt.show()
```

pendataan visualisasi yang terjadi dari tahun terjadinya dan banyaknya kerusakan yang terjadi. dan di data tersebut di tahun 1500-2000 banyak terjadinya kerusakan akibat tsunami.
